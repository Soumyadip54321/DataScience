Importing the necessary modules.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
stu_df=pd.read_csv('student/student-mat.csv')
print(stu_df.head())
print(stu_df.dtypes)
print(stu_df.describe())
```

```python
print(stu_df.columns)
```

Now, we count the frequency of students who hail from urban and rural backgrounds.

```python
print(stu_df.address.value_counts())
```

It'd be better if we could get a relative view of the backgrounds students come from hence we use proportion inlieu of counts.

```python
print(stu_df.address.value_counts(normalize=True))
```

To help visualize the above we plot bar graphs.

```python
sns.countplot(x="address",data=stu_df)
address_list=['Urban','Rural']
plt.title("Bar graph depicting the relative counts of students from various addresses")
plt.legend(address_list)
plt.show()
plt.close()
```

Now,we look at the distribution of absentees over the years.

```python
from scipy.stats import iqr
from scipy.stats import trim_mean
```

```python
print("The interquartile range is:",iqr(stu_df.absences))
print("The range is:",(stu_df.absences.max()-stu_df.absences.min()))
print("The typical value of absence is:",round(stu_df.absences.mean()))
print("The mean absolute deviation is:",round(stu_df.absences.mad()))
print("The most frequent days students remain absent is:",stu_df.absences.mode())
print("The std deviation is:",stu_df.absences.std())
print("The median is:",stu_df.absences.median())
print("The mean with 20% extreme values trimmed is:",trim_mean(stu_df.absences,proportiontocut=0.2))
```

Now,to visualize the above we use "box plot" & histogram both.

```python
sns.boxplot(x="absences",data=stu_df)
#plt.ylabel("days students remains absent")
plt.title("BOX PLOT OF STUDENT ABSENCES OVER THE YEARS")
plt.xlabel("absences in days")
plt.annotate('median @ '+str(stu_df.absences.median()),xy=(stu_df.absences.median(),0),xytext=(stu_df.absences.median(),0.45),arrowprops=dict(facecolor='white',shrink=0.2))
plt.annotate('maximum @'+str(stu_df.absences.max()),xy=(stu_df.absences.max(),0),xytext=(stu_df.absences.max(),0.45),arrowprops=dict(facecolor='red',shrink=0.2))
plt.show()
plt.close()

sns.histplot(x="absences",data=stu_df)
plt.title("HISTOGRAM OF STUDENT ABSENCES OVER THE YEARS")
plt.annotate('median taken cause of skewed histogram located @'+str(stu_df.absences.median()),xy=(stu_df.absences.median(),3),xytext=(stu_df.absences.median(),30),arrowprops=dict(facecolor='orange',shrink=0.8))
plt.show()
plt.close()
```
