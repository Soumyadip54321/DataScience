## This workbook demonstrates usage of EDA towards cleaning a dataset to help build ML models.

```python
import numpy as np
import pandas as pd
import os
```

Display the first few lines of the diabetes dataframe to have a glance at the data content.

```python
os.listdir()
diabetes_df=pd.read_csv('diabetes.csv')
diabetes_df.head()
```

Display the data types of each variables/features and check on the number of observations present.

```python
print(diabetes_df.dtypes)
diabetes_df.info()
```

From the above we see that the 'outcome' column is marked as 'object'/'string' whereas we expected it to be bool/int. This issue could be cause of missing values or other anomalies. To detect let us print all unique values.

```python
diabetes_df.Outcome.unique()
```

As evident the column contains 'O' which must have forced it to become of type object. Now we analyze as to what might have happened.Let's replace the '0' and the change the data type.

```python
diabetes_df.Outcome=diabetes_df.Outcome.replace('O',0)
diabetes_df.Outcome=diabetes_df.Outcome.astype("int64")
diabetes_df.dtypes
```

Compute the summary statistics.

```python
diabetes_df.describe()
```

As can be seen above min values of all 5 cols BMI,insulin,SkinThickness, BloodPressure is 0 which is no way possible indicating that something is off.

Also the max value of Insulin 846 is abnormally high.

Let us replace the missing values in these 5 cols.

```python
diabetes_df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = diabetes_df[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.nan)
```

Let us again check for missing values.

```python
diabetes_df.isnull().sum()
```

Let's print all of the rows that contain missing values.

```python
diabetes_df[diabetes_df.isnull().any(axis=1)]
```
