## This workbook demonstrates the applicability of Data analysis on a dataset.

Here we consider the dataset pertaininig to a fictional online shoe-store and highlights its customer information.

```python
import numpy as np
import pandas as pd
import os
```

```python
os.listdir()
orders=pd.read_csv('shoefly.csv')
```

Now, we check the first few lines of 'SHOEFLY' company's data.

```python
print(orders.head())
print(orders.info())
```

Here we select all the email addresses present for our mkt dept to shoot out emails to.

```python
emails=orders.email
emails
```

Here, we select order of a person named 'francis palmer' since her order was wrong.

```python
frances_palmer=orders[(orders.first_name=='Frances')&(orders.last_name=='Palmer')]
print(frances_palmer)
```

Now, we select those shoes that are deemed comfortable to provide rating to by customers.

```python
comfy_shoes=orders[orders.shoe_type.isin(['clogs','boots','ballet flats'])]
```

Here, we create a seperate column indicating the email provider's names

```python
orders['provider']=orders.email.apply(lambda x:x.split('@')[-1])
orders
```

Now suppose we add a new column the value of which depends on columns that already exists. Here we apply lambda functions on rows that spans multiple columns.

```python
orders['shoe_source']=orders.shoe_material.apply(lambda x:'animal' if x == 'leather' else 'vegan')
orders.head()
```

Now we create a column that indicates salutation to use when shooting mail to respective customers.

```python
orders['salutation']=orders.apply(lambda row:'Dear Mr. '+row['last_name'] if row['gender']=='male' else 'Dear Ms. '+row['last_name'],axis=1)
orders.head()
```
