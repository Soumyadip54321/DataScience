Now we capture the data frame from csv for user_visits to our "SHOEFLY" website.

```python
df=pd.read_csv('user_visits.csv')
df.head()
```

Now we compute number of visits from different utm sources which implies where from users entered our website by clicking on the ads displayed. 

```python
click_source=df.groupby('utm_source').id.count().reset_index()
click_source
```

Suppose we wanna figure out how much the traffic was to our site across each months from various utm_sources.

```python
click_source_by_month=df.groupby(['utm_source','month']).id.count().reset_index()
click_source_by_month=click_source_by_month.rename(columns={
    'id':'traffic'
})
click_source_by_month
```

```python
click_source_by_month_pivot=click_source_by_month.pivot(columns='month',index='utm_source',values='traffic').reset_index()
click_source_by_month_pivot
```
